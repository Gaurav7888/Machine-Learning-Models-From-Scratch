```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
data=datasets.load_iris()
X=data['data']
print(X)

Y=data['target']!=2
Y=Y.astype('int64')
print(Y)
print(X.shape)
print(Y.shape)
```

```python
theta=0
theta=np.ones(X.shape[1])
costly=0
    
    
```

```python
def cost(theta):
    costly=-(Y*(np.log(1/1+np.exp(-(np.dot(X,theta))))) + (1-Y)*(np.log(1-1/1+np.exp(-(np.dot(X,theta))))))/X.shape[0]
    return costly
```

```python
def derv(theta):
    
    derv=np.dot(np.dot(X,theta),  1/1+np.exp(-(np.dot(X,theta))) - Y) 
    derv/=X.shape[0]
    return derv
```

```python
alpha=0.05
iter=1000
cost_=[]
for i in range(iter):
    theta=theta-alpha*derv(theta)
    
    cost_.append(cost(theta))
    
```
