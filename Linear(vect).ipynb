```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from numpy.linalg import norm
```

```python
data = pd.read_csv("50_Startups.csv")
```

```python
data.head()
```

```python
data=data.drop(['State'],axis=1)
```

```python
X=data.iloc[:,0:3]
Y=data['Profit']
m=X.shape[0]
n=X.shape[1]
```

```python
print(m,n)#no. of datapoints,no. of coloumns
st=StandardScaler()
X=st.fit_transform(X)
```

```python
#hypo
theta=np.ones(n)
print(theta)
```

```python

Y_pred=np.dot(X,theta)
```

```python
def cost(theta):
    cost=np.dot(np.dot(X,theta)-Y,np.transpose(np.dot(X,theta)-Y))+.05*norm(theta)
    cost**=2
    cost/=2*m
    return cost
```

```python

cost(theta)
```

```python
def derv(theta):
    
    derv=np.dot(np.transpose(X),np.dot(X,theta)-Y)
    derv/=m
    return derv
```

```python
derv(theta)
```

```python
alpha=0.05
iter=1000
cost_=[]
for i in range(iter):
    theta=theta-alpha*derv(theta)
    
    cost_.append(cost(theta))
    
    
```

```python
cost_
```

```python
plt.plot(cost_,)
```
